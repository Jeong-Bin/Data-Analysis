```python
import pandas as pd

file = './data/player_stat_month.csv'

df_sort = pd.read_csv(file, encoding = 'cp949')

df_sort.head()
```

```python
df_sort.columns
```

```python
df_selected = df_sort[['팀', '이름', '출루율', '03', '04', '05', '06', '07', '08', '09', '10']]
df_selected
```

```python
# heatmap 으로 시각화 할 때 팀, 이름 컬럼 같이 문자가 들어있으면 에러가 남.
# 때문에 팀, 이름 컬럼은 인덱스로 바꿔줌

df_selected = df_selected.set_index(['팀', '이름'])
df_selected
```

```python
# seaborn, matplotlib 의 이미지에서 한글 표현하기
# 외우지말고 그냥 복붙해서 써도 됨

import matplotlib
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import platform
import seaborn as sns

# 이미지에 한글 표시 설정
if platform.system() == 'Windows': # 윈도우인 경우 맑은고딕
    font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
    rc('font', family=font_name)
else :  # Mac인 경우 애플고딕
    rc('font', family='AppleGothic')

# 그래프에서 마이너스 기호가 표시되도록 하는 설정
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
sns.heatmap(df_selected)
```

```python
# heatmap 옵션 추가

fig, ax = plt.subplots(figsize = (10, 10))
sns.heatmap(df_selected,
           annot = True, fmt = '.3f',
           cmap = 'Reds')
```

```python
df_selected.columns[1:]  # 0번째에 있는 전체 출루율은 제외하기 위해
```

```python
# 월별 출루율을 시즌 출루율에 대비한 +/- 값으로 변경하여 월별 변화정도를 살펴보기

# 시즌 전체 대비 월별 실적
for col in df_selected.columns[1:]:
    df_selected[col] = df_selected[col] - df_selected['출루율'] 
    # 해당 월이 + 라면 시즌평균보다 좋았고, - 라면 안 좋음
    # 근데 지금 이상하게 - df_selected['출루율']  이게 적용이 안 됨...
    # 일단 계속 ㄱㄱ

df_selected['출루율'] = 0.0

df_selected.head()
```

```python
fig, ax = plt.subplots(figsize = (10, 10))
sns.heatmap(df_selected,
           annot = True, fmt = '.3f',
           cmap = 'RdBu_r')
```
