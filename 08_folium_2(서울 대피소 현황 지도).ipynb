```python
import folium
```

```python
import pandas as pd
```

```python
# 서울 열린데이터 광장에서 자료 다운

file = './data/서울시 대피소 방재시설 현황 (좌표계_ WGS1984).csv'
raw = pd.read_csv(file, encoding = 'cp949')
# 엑셀과 같은 MS-office 파일은 utf-8(기본값)이 아닌, cp949로 인코딩한다.

raw.head()
```

```python
raw.info()
```

```python
# 인덱스 번호가 0인 경우 위도와 경도 출력

i = 0
lat = raw.loc[i, '위도']
long = raw.loc[i, '경도']
print(lat, long)
```

```python
# 위에서 명칭 추가

i = 0
lat = raw.loc[i, '위도']
long = raw.loc[i, '경도']
name = raw.loc[i, '대피소명칭']
print(lat, long, name)
```

```python
len(raw)  # 데이터 길이(수)
```

```python
# 모든 인덱스 번호에서 출력

for i in range(len(raw)):

    lat = raw.loc[i, '위도']
    long = raw.loc[i, '경도']
    name = raw.loc[i, '대피소명칭']

   # print(lat, long, name)
```

```python
# 마커 추가하기

# 지도 생성
m = folium.Map(location = [37.5536067, 126.9674308],
               zoom_start = 12)

# 마커 추가
for i in range(len(raw)):

    lat = raw.loc[i, '위도']
    long = raw.loc[i, '경도']
    name = raw.loc[i, '대피소명칭']
    
    folium.Marker([lat, long],
                 tooltip = name).add_to(m)

m
```

```python
# 지도 저장하기(새로운 folium에선 툴팁 한글이 깨져서 html 웹 페이지로 봐야함)

m.save('./data/Sheltermap.html')
```

```python
# MarkerCluster

from folium.plugins import MarkerCluster
```

```python
# MarkerCluster 로 근처에 있는 마커들 그룹으로 묶기

# 지도 생성
m = folium.Map(location = [37.5536067, 126.9674308],
               zoom_start = 12)

marker_cluster = MarkerCluster().add_to(m)

# 마커 추가
for i in range(len(raw)):

    lat = raw.loc[i, '위도']
    long = raw.loc[i, '경도']
    name = raw.loc[i, '대피소명칭']
    
    folium.Marker([lat, long],
                 tooltip = name).add_to(marker_cluster)

m
```
