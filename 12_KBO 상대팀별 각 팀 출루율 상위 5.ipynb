```python
import pandas as pd
```

```python
file = './data/KBO_2019_player_gamestats.csv'
raw = pd.read_csv(file, encoding = 'cp949')
raw.head()
```

```python
raw['상대'].unique()
```

```python
# @ 는 상대방 홈구장 경기.

opp_list = [ ]
home_away_list = [ ]

for opp in raw['상대']:
    
    if '@' in opp:
        home_away = '원정'
        opp = opp.replace('@', '')  # @ 없애서 상대팀명을 통일시킴
    else:
        home_away = '홈'
        
    opp_list.append(opp)
    home_away_list.append(home_away)
    

# 컬럼 추가
raw['홈어웨이'] = home_away_list
raw['상대팀'] = opp_list

raw.head()
```

```python
home_away_list
```

```python
raw.head()
```

```python
data = raw.pivot_table(index = ['팀', '이름', '생일', '상대팀'],
                       values = ['타수', '안타', '홈런', '루타', '타점', '볼넷', '사구', '희비'],
                       aggfunc = 'sum')

data.head()
```

```python
# 타수가 0보다 큰 사람만 보기

cond = data['타수'] > 0
data = data[cond]  # data라는 동일한 이름으로 다시 저장
data.head()
```

```python
# index를 리셋해서 컬럼으로 만들고 다시 data에 저장
data = data.reset_index()
data.head()
```

```python
def cal_hit(df):
    df['타율'] = df['안타'] / df['타수']
    df['출루율'] = (df['안타'] + df['볼넷'] + df['사구']) / (df['타수'] + df['사구'] + df['희비'])
    df['장타율'] = df['루타'] / df['타수']
    df['OPS'] = df['출루율'] + df['장타율']
    
    return df
```

```python
player_stats_opp = cal_hit(data)
player_stats_opp.head()
```

```python
# 특정팀을 상대로 가장 강한 타자. 단 타수가 10 넘는 경우

team = '두산'
cond = (player_stats_opp['상대팀'] == team) & (player_stats_opp['타수'] > 10)

# 출루율 상위 10
player_stats_opp[cond].sort_values(by = '출루율', ascending = False).head(10)
```

```python
team = '롯데'
cond = (player_stats_opp['상대팀'] == team) & (player_stats_opp['타수'] > 20)

# 출루율 상위 10
player_stats_opp[cond].sort_values(by = '출루율', ascending = False).head(10)
```

```python
# KBO 전체 팀을 대상으로 팀별 출루율 상위 5인

hitter_df = pd.DataFrame() # 기본 빈 데이터프레임을 hitter_df에 넣음

for team in player_stats_opp['상대팀'].unique():
    
    cond = (player_stats_opp['상대팀'] == team) & (player_stats_opp['타수'] > 20)
    df = player_stats_opp[cond].sort_values(by = '출루율', ascending = False).head(5)
    
    hitter_df = hitter_df.append(df)
```

```python
hitter_df
```

```python
# unique()를 이용해 복수의 팀을 대상으로 탑5에 들었을 경우 한 번만 나타나게 함

hitter_df['이름'].unique()
```

```python
# 위 데이터의 타자들을 대상으로 팀별 출루율 피벗테이블 만들기

cond = [ ]

for name in player_stats_opp['이름']:
    
    if name in hitter_df['이름'].unique():
        cond.append(True)
    else:
        cond.append(False)
        
player_stats_opp[cond]

# True에 해당하는 데이터만 가져옴
```

```python
# 다른 방식. isin을 이용한 조건문

cond = player_stats_opp['이름'].isin( hitter_df['이름'].unique() )
top_df = player_stats_opp[cond]

top_pivot = top_df.pivot_table(index = ['팀', '이름'], values = '출루율', columns = '상대팀')
top_pivot
```
