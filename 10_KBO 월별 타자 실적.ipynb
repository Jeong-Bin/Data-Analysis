```python
# KBO 월별 타자 실적

import pandas as pd

file = './data/KBO_2019_player_gamestats.csv'

raw = pd.read_csv(file, encoding = 'cp949')

raw.head()
```

```python
raw.columns
```

```python
# 일자 컬럼에서 월만 추출하기

month_list = [ ] # 2. 아래의 데이터를 여기에 저장

for monthdate in raw['일자']:
    month = monthdate.split('-')[0]  # 1. - 를 기준으로 분리 후 첫 번째 원소 지정
    month_list.append(month)  # 3. month_list에 month 데이터 추가
    
raw['월'] = month_list

raw.head()
```

```python
# 분석에 활용할 컬럼만 선택

columns_select = ['팀', '이름', '생일', '일자', '상대', '타수', '안타', '홈런', '루타', '타점', '볼넷', '사구', '희비', '월']
data = raw[columns_select]

data.head()
```

```python
# 피벗테이블로 월별 실적 집계

data_player_month = data.pivot_table(index = ['팀', '이름', '생일', '월'],
                    values = ['타수', '안타', '홈런', '루타', '타점', '볼넷', '사구', '희비'],
                    aggfunc = 'sum')

data_player_month.head()
```

```python
data_player_month.info()
```

```python
# 팀,이름,생일,월 로 정리된 index를 다시 컬럼으로 변경

data_player_month = data_player_month.reset_index()
data_player_month.head()
```

```python
# 함수로 계산

def cal_hit(df):
    df['타율'] = df['안타'] / df['타수']
    df['출루율'] = (df['안타'] + df['볼넷'] + df['사구']) / (df['타수'] + df['볼넷'] + df['사구'] + df['희비'])
    df['장타율'] = df['루타'] / df['타수']
    df['OPS'] = df['출루율'] / df['장타율']
    
    return df

cal_hit(data_player_month)
```

```python
player_month_stat = cal_hit(data_player_month)
player_month_stat.info()

# 타율, 출루율, 장타율의 데이터 수가 적은 것을 확인할 수 있음
```

```python
player_month_stat.dropna()
# dropna()를 하면 어느 하나라도 데이터가 비어있으면, 해당 행을 삭제함.
```

```python
player_month_stat = player_month_stat.dropna()
player_month_stat.head()
```

```python
month_pivot = player_month_stat.pivot_table(index = ['팀', '이름', '생일'],
                                          values = '출루율',
                                          columns = '월')
month_pivot.head()
```

```python
# 이전에 만들어둔 타자 실적 데이터를 불러옴

file = './data/player_stat.csv'
player_stat = pd.read_csv(file, encoding = 'cp949')
player_stat.head(10)
```

```python
# 타자 실적과 월별 출루율 데이터를 병합함
# left_on = ['팀', '이름', '생일'], right_on = ['팀', '이름', '생일']
# 이렇게 좌우 값이 같다면 아래와 같이 on 만 써도 된다.
df = pd.merge(player_stat, month_pivot, how = 'left', on = ['팀', '이름', '생일'] )
df.head()
```

```python
# 이번엔 출루율 실적을 기준으로 정렬

df_sort = df.sort_values(by = '출루율', ascending = False).head(50)
df_sort
```

```python
file = './data/player_stat_month.csv'

df_sort.to_csv(file, encoding = 'cp949', index = False)
```
