```python
import os
```

```python
dirpath = './data/subway_data/'

files = os.listdir(dirpath) # 해당 폴더 안에 어떤 파일들이 있는지 보여줌
files
```

```python
# 파일 병합하는 과정

# 1. 빈 데이터프레임 준비
raw = pd.DataFrame()

# 2. 반복문 -> csv 파일 읽기 -> 데이터프레임에 추가
for file in files:
    temp = pd.read_csv(dirpath + file)  # 경로와 함께 넣어줘야함
    raw = raw.append(temp)
    # raw = raw.append(temp, ignore_index = True)  4) 아래 방법들 말고 이렇게 해도 됨.

raw.head()
```

```python
# 요일 추가하기

from datetime import datetime
```

```python
# datetime.strptime('날짜str',str형태):문자 -> 날짜 타입
#     str형태 : %Y-%m-%d %H:%M:%S
#     %Y : 연도(4자리)
#     %m : 월(2자리)
#     %d : 일자(2자리)
#     %H : 시
#     %M : 분
#     %S : 초

# 예시)
# s = '2019-01-05'
# datetime.strptime(s, '%Y-%m-%d')

s = '20190630'   # 만약 - 가 없으면 아래에도 -를 빼준다
date = datetime.strptime(s, '%Y%m%d')
date
```

```python
# 날짜타입으로 요일을 변경할 때는 weekday()를 사용한다.
# 월요일 : 0 ~ 일요일 : 6

# 예시)
# 0 ~ 6 으로 표기된 요일을 list 에 넣어서 다시 문자로 바꿈
weekday_dict = ['월', '화', '수', '목', '금', '토', '일']
weekday_index = date.weekday()
print(weekday_index)  # 6
print(weekday_dict[weekday_index])  # 일
```

```python
# 날짜 컬럼을 불러와서 순서대로 요일을 점검하여 리스트에 저장하기

weekday_dict = ['월', '화', '수', '목', '금', '토', '일'] # 4)

weekday_list = [ ] # 6)

for date_str in raw['사용일자']:  # 1)
    date = datetime.strptime(str(date_str),'%Y%m%d') # 2) # date_str을 문자열 형태로 바꿔줘야 함.
    weekday_index = date.weekday() # 3)
    weekday = weekday_dict[weekday_index] # 5)
    
    weekday_list.append(weekday)  # 7)
    
```

```python
raw['요일'] = weekday_list

raw.head()
```
